```python
with open(output_dir + 'SYB_Series_Catalog.json') as json_file:  
    catalogue = json.load(json_file)


```

```python
def read_data_file(topic, table, series, dir_path):
    file = 'wide_Topic'+topic+'_Table'+table+'_Series'+series+'.json'
    with open(dir_path + file) as json_file:  
        data = json.load(json_file)
    return data
```

```python
data = read_data_file(topic = '02', table = '008', series = 'SYB011', dir_path = output_dir)
data.keys()

data_header = {k: data[k] for k in data.keys() if k in ['topicId', 'topicNameEN', 'topicNameFR',
                                          'tableCode', 'tableId', 'tableName', 
                                          'seriesDesc', 'seriesCode', 'SYB_series']}

data_header

```

```python
for t1 in catalogue:
    for t2 in t1['tables']:
        for s in t2['series']:
            #if s['seriesCode'] != 'SYB012':
            #    continue
                
            topic = str(t1['topicId']).zfill(2)
            table = str(t2['tableId']).zfill(3) 
            series = s['seriesCode']
            
            file_name = 'wide_Topic'+ topic + '_Table' + table + '_Series' + series + '.json'


            data = read_data_file(topic = str(t1['topicId']).zfill(2), 
                                  table = str(t2['tableId']).zfill(3), 
                                  series = s['seriesCode'], 
                                  dir_path = output_dir)
            
            #data['data'][0]['data'][0]
            
            matrix = []
            
            for g in data['data']:
                for r in g['data']:
                    d={}
                    d['topicId'] = data['topicId']
                    d['topicNameEN'] = data['topicNameEN']
                    d['topicNameFR'] = data['topicNameFR']
                    d['tableCode'] = data['tableCode']
                    d['tableId'] = data['tableId']
                    d['tableName'] = data['tableName']
                    d['seriesDesc'] = data['seriesDesc']
                    d['seriesCode'] = data['seriesCode']
                    d['parentCode'] = g['parentRegionId']
                    d['parentName'] = g['parentRegion_DescEN']
                    d['countryProfile'] = g['countryProfile']
                    d['refAreaCode'] = g['refAreaCode']
                    d['refAreaDesc'] = g['refAreaDesc']
                    d['ISO3'] = g['ISO3']
                    d['UN_Member'] = g['UN_Member']
                    d['X'] = g['X']
                    d['Y'] = g['Y']
                    for k,v in r.items():
                        d[k] = v
                    matrix.append(d)
                    
            if len(matrix)>0:
                keys = matrix[0].keys()
            
                #keys

                


                file = 'csv_Topic'+topic+'_Table'+table+'_Series'+series+'.csv'
                
                print('printing ' + file + '\n')

                with open(output_dir + file, 'w', newline='', encoding='utf-8') as csv_file:  
                    dict_writer = csv.DictWriter(csv_file, keys, delimiter='\t')
                    dict_writer.writeheader()
                    dict_writer.writerows(matrix)


```
